```python
import pandas as pd
```

```python
data_folder1 = 'Data/mturkfitbit_export_3.12.16-4.11.16/Fitabase Data 3.12.16-4.11.16/'
data_folder2 = 'Data/mturkfitbit_export_4.12.16-5.12.16/Fitabase Data 4.12.16-5.12.16/'
```

```python
df_activity_daily_part1 = pd.read_csv(data_folder1 + 'dailyActivity_merged.csv', header=0)
df_activity_daily_part2 = pd.read_csv(data_folder2 + 'dailyActivity_merged.csv', header=0)
```

```python
df_activity_daily_part1.head(10)
```

```python
df_activity_daily_part2.head(10)
```

```python
df_activity_daily = pd.concat([df_activity_daily_part1, df_activity_daily_part2])
```

```python
df_activity_daily.describe()
```

```python
df_activity_daily['ActivityDate'] = pd.to_datetime(df_activity_daily['ActivityDate'], format='%m/%d/%Y')
```

```python
df_activity_daily.sort_values('ActivityDate', ascending= True)
```

```python
import os
output_folder = "output/activity/"
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# 2. 按照日期分组并循环
# 注意：确保 ActivityDate 已经是 datetime 类型，否则请先执行 pd.to_datetime
for date, group in df_activity_daily.groupby('ActivityDate'):
    
    # 将日期转为字符串作为文件名 (例如: 2016-04-12.csv)
    file_name = f"activity_{date.strftime('%Y-%m-%d')}.csv"
    file_path = os.path.join(output_folder, file_name)
    # 3. 输出为 CSV
    # index=False 避免保存 pandas 的行索引
    group.to_csv(file_path, index=False)
    
    print(f"已导出: {file_name}")

print(f"\n✅ 所有文件已成功保存至目录: {output_folder}")
```
